```python

# Valores de las resistencias (en kΩ)
R1 = 0.2778
R2 = 1.6200
R3 = 0.5700
R4 = 9.9470

# Incertidumbres (en kΩ)
u_R1 = 0.0020
u_R2 = 0.0081
u_R3 = 0.0036
u_R4 = 0.0937

# Función de la resistencia equivalente
def Req(R1, R2, R3, R4):
    return R1 + 1 / (1/R3 + 1/(R2 + R4))

# Valor de la resistencia equivalente
Req_val = Req(R1, R2, R3, R4)

# Derivadas parciales
den = (1/R3 + 1/(R2 + R4))**2

dReq_dR1 = 1
dReq_dR2 = 1 / ((R2 + R4)**2 * den)
dReq_dR3 = 1 / (R3**2 * den)
dReq_dR4 = 1 / ((R2 + R4)**2 * den)

# Propagación de incertidumbres
u_Req = np.sqrt(
    (dReq_dR1 * u_R1)**2 +
    (dReq_dR2 * u_R2)**2 +
    (dReq_dR3 * u_R3)**2 +
    (dReq_dR4 * u_R4)**2
)

print(f"Req = {Req_val:.4f} kΩ")
print(f"u(Req) = {u_Req:.4f} kΩ")
print(f"Req = ({Req_val:.4f} ± {u_Req:.4f}) kΩ")
```

### Kirchoff

- Regla de los nudos: La suma de las corrientes que entran en un nudo es igual a la
suma de las corrientes que salen del nudo.
- Regla de las mallas: La suma de todas las variaciones de tensión (caídas y subidas)
en una malla es igual a cero.

```python
#Nudos
Nudo={"Voltios":[5,5.5,6,6.5,7],
      "R1":[6.36,6.93,7.63,8.27,8.86],
      "R3":[-6.04,-6.67,-7.31,-7.95,-8.42],
      "R2":[-0.29,-0.31,-0.34,-0.37,-0.4]
      }

df_Nudo = pd.DataFrame(Nudo)
df_Nudo["Corriente neta"]=df_Nudo.apply(lambda row: sum(row[1:]), axis=1)





print(df_Nudo)
print(df_Nudo.to_latex(index=False, float_format="%.2f", caption="Voltajes en el nudo", label="tab:nudo"))
```
